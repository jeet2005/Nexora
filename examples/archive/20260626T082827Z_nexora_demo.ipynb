# 🚀 Nexora — Autonomous Predictive Analytics

**From CSV to trained model in one line of code.**

This notebook demonstrates the full Nexora workflow:
1. Install the package
2. Load data
3. Profile the dataset
4. Train models automatically
5. View the leaderboard
6. Make predictions

## 📦 Step 1: Install Nexora

## 📥 Step 2: Import & Load Data

```python
import pandas as pd

from nexora import Nexora

# Option A: Load from a CSV file
# nx = Nexora("your_data.csv", target="price")

# Option B: Load from sklearn datasets (built-in)
nx = Nexora.from_sklearn("california_housing", target="MedHouseVal")

print(nx)
```

## 🔍 Step 3: Profile the Dataset

Get a comprehensive health check of your data — missing values, distributions, correlations, and more.

```python
profile = nx.profile()

print(f"📊 Dataset: {profile.source_name}")
print(f"📏 Shape: {profile.num_rows:,} rows × {profile.num_columns} columns")
print(f"🎯 Target: {profile.target}")
print(f"📈 Task Type: {profile.task_type}")
print(f"💚 Health Score: {profile.health_score:.1f}/100")
print(f"🔢 Numeric features: {profile.num_numeric}")
print(f"🏷️ Categorical features: {profile.num_categorical}")
print(f"❓ Missing values: {profile.total_missing}")
```

## 🤖 Step 4: Train Models (One Line!)

Nexora automatically:
- Detects task type (regression/classification)
- Preprocesses data (scaling, encoding, missing values)
- Trains multiple models (LinearRegression, RandomForest, XGBoost, LightGBM, etc.)
- Ranks them by performance

```python
# Full training — trains up to 6 models
report = nx.run()

# Or for a quick 30-second run:
# report = nx.quick()
```

## 🏆 Step 5: View the Leaderboard

```python
# See all trained models ranked by performance
report.leaderboard()
```

```python
# Best model details
best = report.best_model
print(f"🏅 Best Model: {best.model_name}")
print(f"📊 Primary Metric ({best.primary_metric}): {best.primary_score:.4f}")
print(f"⏱️ Training Time: {best.train_time_sec:.2f}s")
print("\n📈 All Metrics:")
for metric, value in best.metrics.items():
    print(f"   {metric}: {value:.4f}")
```

## 🔮 Step 6: Make Predictions

```python
# Predict on new data (using first 5 rows as example)
sample = nx.df.drop(columns=["MedHouseVal"]).head(5)
predictions = report.predict(sample)

print("🔮 Predictions:")
for i, pred in enumerate(predictions):
    actual = nx.df["MedHouseVal"].iloc[i]
    print(f"   Row {i + 1}: Predicted = {pred:.4f}, Actual = {actual:.4f}")
```

## 💾 Step 7: Save & Load Your Model

```python
# Save the entire session (model + metadata + profiling)
report.save("my_model.nx")
print("✅ Model saved to my_model.nx")

# Load it back later
loaded_report = Nexora.load("my_model.nx")
print(f"🔄 Loaded: {loaded_report.best_model.model_name}")
```

## 📊 Step 8: Classification Example

```python
# Nexora auto-detects classification tasks!
nx_cls = Nexora.from_sklearn("iris", target="target")
report_cls = nx_cls.run()

print(f"🎯 Task: {report_cls.task_type}")
report_cls.leaderboard()
```

## 📚 Quick Reference

| Command | Description |
|---------|-------------|
| `Nexora(data, target="col")` | Create session |
| `nx.profile()` | Dataset health check |
| `nx.run()` | Train all models |
| `nx.quick()` | Fast 30s training |
| `report.leaderboard()` | Model rankings |
| `report.best_model` | Best model info |
| `report.predict(data)` | Make predictions |
| `report.save("file.nx")` | Save session |
| `Nexora.load("file.nx")` | Load session |
